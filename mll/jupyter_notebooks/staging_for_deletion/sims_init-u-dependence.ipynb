```python
import pandas as pd
```

```python
sim_db = '~/sims_n_results/sfcompo_compare_aug2020/van2_15nuc_plusU_4.pkl'
XY = pd.read_pickle(sim_db)
XY.reset_index(inplace=True, drop=True)
```

```python
labels = ['AvgPowerDensity', 'Burnup', 'CoolingTime', 'Enrichment', 
          'ModDensity', 'OrigenReactor', 'ReactorType', 'UiWeight']
```

```python
XY = XY.loc[XY.CoolingTime == 0.0].copy()
```

```python
XY
```

```python
isos = XY.columns[~XY.columns.isin(labels)].tolist()
print(isos)
```

```python
XY[isos] = XY[isos].div(XY['total'], axis=0)
XY.drop('total', axis=1, inplace=True)
```
