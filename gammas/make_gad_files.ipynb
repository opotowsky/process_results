```python
import pandas as pd
```

```python
db = pd.read_pickle('not-scaled_nuc32_activities.pkl')
```

```python
lbls = ['ReactorType', 'CoolingTime', 'Enrichment', 'Burnup', 'OrigenReactor']
nonlbls = ['AvgPowerDensity', 'ModDensity', 'UiWeight']
nuc_list = ['ac227', 'am241', 'am243', 'ba133', 'cf249', 'cf252', 'cm243', 'cm244', 
            'cm245', 'cs134', 'cs137', 'eu152', 'eu154', 'ho166m', 'kr85', 'nb94', 
            'np236', 'np237', 'pa231', 'pm146', 'pu236', 'pu238', 'pu239', 'pu240', 
            'ra226', 'sb125', 'th228', 'th229', 'u232', 'u233', 'u234', 'u235']
```

```python
db.shape
```

```python
db.reset_index(drop=True, inplace=True)
```

```python
db.head()
```

```python
# don't rerun this cell
# scaling all activities to be from 1MT material to 1g so gadras detector setups can handle the sources
db[nuc_list] = db[nuc_list]/1e6
```

```python
db.head()
```

```python
db.to_pickle('nuc32_activities_scaled_1g_reindex.pkl')
```

```python
# source age for gadras needs to be in years, converting from cooling time in days
#db['Age'] = (1 / 365.25) * db['CoolingTime'].copy()

# actually, age needs to be above but close to zero because these activities are pre-decayed.
# study is in a directory in gadras detector response repo
# the below time is 20 min
db['Age'] = 3.805175e-5 
db.head()
```

```python
db = db[['Age'] + nuc_list]
db.head()
```

```python
db.to_csv('nuc32.act', sep = ' ', header=False)
```
