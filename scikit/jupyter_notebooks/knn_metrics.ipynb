```python
from sklearn.preprocessing import scale
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score

import pandas as pd
import numpy as np
```

```python
from learn.tools import splitXY, random_error
```

```python
train = '~/sims_n_results/final_sims_nov2020/not-scaled_nuc29.pkl'
trainXY = pd.read_pickle(train)
trainXY.reset_index(inplace=True, drop=True)
trainXY = trainXY.sample(frac=0.05)
trainX_unscaled, rY, cY, eY, bY = splitXY(trainXY)
```

# Reactor

```python
trainY = rY
k = 2
alg = 'knn'
CV = 3
kfold = StratifiedKFold(n_splits=CV, shuffle=True)
scores = 'accuracy'
rxtr_param = 'reactor'
```

## Minkowski, p=1, 2, 5, 10

```python
init = KNeighborsClassifier(n_neighbors=k, weights='distance', p=1, metric='minkowski')
csv_name = 'mink1'
random_error(trainX_unscaled, trainY, alg, init, kfold, scores, csv_name, rxtr_param)
mink = pd.read_csv(csv_name + '_random_error.csv')
mink
```

```python
init = KNeighborsClassifier(n_neighbors=k, weights='uniform', p=1, metric='minkowski')
csv_name = 'mink1'
random_error(trainX_unscaled, trainY, alg, init, kfold, scores, csv_name, rxtr_param)
mink = pd.read_csv(csv_name + '_random_error.csv')
mink
```

```python
init = KNeighborsClassifier(n_neighbors=k, weights='distance', p=2, metric='minkowski')
csv_name = 'mink2'
random_error(trainX_unscaled, trainY, alg, init, kfold, scores, csv_name, rxtr_param)
euc = pd.read_csv(csv_name + '_random_error.csv')
euc
```
